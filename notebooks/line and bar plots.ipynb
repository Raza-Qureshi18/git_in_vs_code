# Line plot

line plot is always drawn between two variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark",color_codes=True)
kashti = sns.load_dataset("iris")
sns.lineplot( x="petal_length",y= "sepal_length",data=kashti,color="red")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark",color_codes=True)
kashti = sns.load_dataset("iris")
sns.lineplot( x="petal_length",y= "sepal_length",data=kashti,color="red")
plt.title("line plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid",color_codes=True)
kashti = sns.load_dataset("iris")
sns.lineplot( x="petal_length",y= "sepal_length",data=kashti,color="red")
plt.title("line plot")
plt.xlim(5)
plt.ylim(5)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark",color_codes=True)
kashti = sns.load_dataset("iris")
plt.figure(figsize=(4,4))
sns.lineplot( x="petal_length",y= "sepal_length",data=kashti,color="red")
plt.title("line plot")
plt.show()
```

# bar plot

it has one catagorical value and one numerical value.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark",color_codes=True)
kashti = sns.load_dataset("iris")
plt.figure(figsize=(4,4))
sns.barplot( x="species",y= "sepal_length",data=kashti)
plt.title("line plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark",color_codes=True)
kashti = sns.load_dataset("titanic")
sns.barplot( x="sex",y= "alone",hue="class",data=kashti,saturation=0.3,color="teal")
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
sns.barplot(x="sex",y="alone",hue="class",errcolor="none",data=titanic,edgecolor="0.7",facecolor=(1,1,1,0),linewidth=3)
plt.show
```
