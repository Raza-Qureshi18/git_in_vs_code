### bar plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
kashti = sns.load_dataset("titanic")
sns.barplot( x= "sex",y= "survived",data=kashti)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark",color_codes=True)
kashti = sns.load_dataset("titanic")
sns.catplot( x= "sex",y= "survived",hue="class",kind="bar",data=kashti,saturation=0.3)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=False)
kashti = sns.load_dataset("titanic")
sns.catplot( x= "sex",y= "survived",hue="class",kind="bar",data=kashti,saturation=0.3)
plt.show()
```

# Count plot

In countplot the one of the axis is allways count itself.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=False)
kashti = sns.load_dataset("titanic")
sns.countplot( x= "sex",data=kashti,saturation=0.3)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark",color_codes=True)
kashti = sns.load_dataset("titanic")
sns.countplot( x= "sex",hue="class",data=kashti,saturation=0.3)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark",color_codes=True)
kashti = sns.load_dataset("titanic")
sns.countplot( x= "sex",hue="class",order=("female","male"),data=kashti,saturation=0.3,color="red")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark",color_codes=True)
kashti = sns.load_dataset("titanic")
sns.countplot( y= "sex",hue="class",order=("female","male"),data=kashti,saturation=0.3,color="orange")
plt.show()
```

# scatter plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
kashti = sns.load_dataset("titanic")
g1=sns.FacetGrid( row= "sex",hue="alone",data=kashti)
g1=(g1.map(plt.scatter,"age","fare").add_legend())
plt.show()
```
