# box plot

it has one numerical value and one catagorical value.

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_style(style="darkgrid")
titanic=sns.load_dataset("titanic")
sns.boxplot(x="class",y="fare",data=titanic)
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_style(style="darkgrid")
titanic=sns.load_dataset("tips")
sns.boxplot(x="day",y="tip",data=titanic)
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_style(style="darkgrid")
titanic=sns.load_dataset("tips")
sns.boxplot(x="tip",y="day",data=titanic)
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_style(style="darkgrid")
titanic=sns.load_dataset("tips")
sns.boxplot(x="tip",y="day",hue="smoker",palette=("red","blue"),data=titanic)
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_style(style="darkgrid")
titanic=sns.load_dataset("tips")
sns.boxplot( y="tip" , x="day" ,showmeans = True, meanprops={"marker":"*","markeredgecolor":"red","markersize":"20"} , data=titanic)
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_style(style="darkgrid")
titanic=sns.load_dataset("tips")
sns.boxplot( y="tip" , x="day" ,showmeans = True, meanprops={"marker":"*","markeredgecolor":"red","markersize":"10"} , data=titanic)
plt.xlabel("days of week")
plt.ylabel("tip given in days")
plt.title("tips graph")
plt.show()
```
